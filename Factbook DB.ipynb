```python
import pandas as pd
import sqlite3 as sql
```

```python
database = "factbook.db"
conn = sql.connect(database)
```

```python
def read_query(q):
    return pd.read_sql_query(q, conn)
```

```python
q = """select * from sqlite_master where type = "table" """

read_query(q)
```

```python
q = """select * from  facts"""

read_query(q)
```

```python
q = """select * from  cities"""

read_query(q)
```

```python
q = """select * from  facts where id = 184"""

read_query(q)
```

```python
q = """select * from  facts where name = "Pakistan" """

read_query(q)
```

```python
q = """select * from  cities where facts_id = 132 """

read_query(q)
```

```python
q = """select * from facts f 
        inner join cities c
        on
        f.id == c.facts_id
        
        where
        f.name = 'Pakistan'"""

read_query(q)
```

```python
q = """select f.name, c.name from facts f 
        inner join cities c
        on
        f.id == c.facts_id
        
        where
        f.name = 'Pakistan'"""

read_query(q)
```

```python
q = """select f.name, c.name, c.capital from facts f 
        inner join cities c
        on
        f.id == c.facts_id
        
        where
        f.name = 'Pakistan' and c.capital == 1"""

read_query(q)
```

```python
q = """select f.name, c.name from facts f 
        inner join cities c
        on
        f.id == c.facts_id
        
        where
        c.capital == 1"""

read_query(q)
```

```python
q = """select f.name, c.name from facts f 
        left join cities c
        on
        f.id == c.facts_id
        """

read_query(q)
```

```python
q = """select f.name Country, c.name Capital, 
        c.population Capital_Population 
        from facts f 
        inner join cities c
        on
        f.id == c.facts_id
        
        where
        c.capital == 1 and c.population > 10000000 """

read_query(q)
```

```python
q = """select f.name Country, f.population Top_pop,
        sum(c.population) Urban_Pop
        from facts f 
        inner join cities c
        on
        f.id == c.facts_id
        group by 1
        """

read_query(q)
```

```python
q = """select f.name Country, f.population Top_pop,
        sum(c.population) Urban_Pop, round((cast(sum(c.population)as float)/f.population)*100, 2) Ur_Population_Percentage
        from facts f 
        inner join cities c
        on
        f.id == c.facts_id
        group by 1
        """

read_query(q)
```
