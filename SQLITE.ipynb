```python
import pandas as pd
import sqlite3 as sql
```

```python
database="jobs.db"
conn=sql.connect(database)
```

```python
pd.read_sql_query("select * from sqlite_master", conn)
```

```python
q="select * from recent_grads"

pd.read_sql_query(q, conn)
```

```python
q="select * from recent_grads limit 5"

pd.read_sql_query(q, conn)
```

```python
q="""select ShareWomen, Major, Major_category from recent_grads 
    where ShareWomen>0.5 and ShareWomen<0.8 limit 10"""

pd.read_sql_query(q, conn)
```

```python
q="""select ShareWomen, Major, Major_category from recent_grads 
    where ShareWomen>0.5 or 
    (ShareWomen<=0.5 and Major_category=="Engineering")"""

pd.read_sql_query(q, conn)
```

```python
q="""select count(Major_category)
    from recent_grads
    where Major_category=="Engineering" """

pd.read_sql_query(q, conn)
```

```python
q="""select distinct(Major_category)
    from recent_grads """

pd.read_sql_query(q, conn)
```

```python
q="""select Major_category, avg(ShareWomen)
    from recent_grads
    group by Major_category 
    """

pd.read_sql_query(q, conn)
```

```python
q="""select Women, Total, round(Women/cast(total as float)*100,2) Women_Percentage from recent_grads"""

pd.read_sql_query(q, conn)
```

```python
q="""select Men, Women, Total, round(Women/cast(total as float)*100,2) Women_Percentage from recent_grads 
    where Women_Percentage>100"""

pd.read_sql_query(q, conn)
```

```python
q="""select Men, Women, (Men + Women) Total_Corr , round(Women/cast(Total as float)*100,2) 
    Women_Percentage from recent_grads 
    """

pd.read_sql_query(q, conn)
```

```python
q="""select Men, Women, (Men + Women)Total_Corr , round(Women/cast(Men+Women as float)*100,2) 
    Women_Percentage, Major from recent_grads 
    where Women_Percentage > 80 """

pd.read_sql_query(q, conn)
```

```python
q = """select Major_category, major, ShareWomen 
        from recent_grads
        where
        ShareWomen > (select avg(ShareWomen) from recent_grads) """

pd.read_sql_query(q, conn) 
```

```python
q = """select Major_category, major, ShareWomen 
        from recent_grads
        where
        ShareWomen > (select avg(ShareWomen) from recent_grads)
        order by 3 DESC
        limit 5"""

pd.read_sql_query(q, conn) 
```

```python
q = """select count(*) 
        from recent_grads
        where
        ShareWomen > (select avg(ShareWomen) from recent_grads) """

pd.read_sql_query(q, conn) 
```

```python
q = """select count(*) from recent_grads"""

pd.read_sql_query(q, conn) 
```

```python
q = """select count(*) No_of_fields_avg_SH,
        (select count(*) from recent_grads) Total_of_fields
        from recent_grads
        where
        ShareWomen > (select avg(ShareWomen) from recent_grads) """

pd.read_sql_query(q, conn) 
```

```python
q = """select count(*) No_of_fields_avg_SH,
        (select count(*) from recent_grads) Total_of_fields,
        round((cast(count(*) as float)/(select count(*) from recent_grads))*100, 2) Percentage
        from recent_grads
        where
        ShareWomen > (select avg(ShareWomen) from recent_grads) """

pd.read_sql_query(q, conn) 
```
